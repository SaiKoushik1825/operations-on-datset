```python
import pandas as pd
import numpy as np

# ... (Your existing code to load the DataFrame) ...

# Check if 'price' or 'Price' column exists in the DataFrame
price_column = 'price' if 'price' in df.columns else 'Price' if 'Price' in df.columns else None

if price_column:
    # Apply np.log1p on prices (safe for 0 values)
    df['log_price'] = np.log1p(df[price_column])
    df[[price_column, 'log_price']].head()
else:
    print("Column 'price' or 'Price' not found in DataFrame")
```

```python
# Check for nulls
df.isnull().sum()

# Drop rows with missing values
df_cleaned = df.dropna()

# Fill missing values
df_filled = df.fillna({'price': 0})
```

```python
# Check which rows have null prices, handling potential column name variations
if 'price' in df.columns:
    null_price_rows = df[df['price'].isnull()]
elif 'Price' in df.columns:
    null_price_rows = df[df['Price'].isnull()]
else:
    print("No 'price' or 'Price' column found for null check")
    null_price_rows = pd.DataFrame()  # Empty DataFrame if column not found

print(null_price_rows)
# Replace null product names with "Unknown"
# Check if 'product_name' or 'Product Name' column exists before filling NaNs
product_name_col = 'product_name' if 'product_name' in df.columns else 'Product Name' if 'Product Name' in df.columns else None

if product_name_col:
    df[product_name_col] = df[product_name_col].fillna("Unknown")
else:
    print("Column 'product_name' or 'Product Name' not found in DataFrame")
```
